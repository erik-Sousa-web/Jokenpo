````python
import pandas as pd
# Importamos a biblioteca Pandas e damos o apelido de 'pd' (padrﾃ｣o no mercado)
import pandas as pd

# O comando read_csv abre o arquivo e o transforma em um 'DataFrame' (uma tabela inteligente)
df = pd.read_csv('SDW2023.csv')

# Exibimos o conteﾃｺdo da tabela para verificar se a leitura deu certo
print("--- DADOS EXTRAﾃ好OS ---")
print(df)
````

````python
# Criamos uma 'funﾃｧﾃ｣o' (um pequeno robﾃｴ) que recebe um nome e devolve uma frase pronta
def criar_frase_marketing(nome_do_cliente):
    # Usamos f-string (o 'f' antes das aspas) para encaixar o nome dentro do texto
    return f"Olﾃ｡ {nome_do_cliente}! O Santander quer ver seu patrimﾃｴnio crescer. Vamos investir?"

# Criamos uma nova coluna chamada 'News' na nossa tabela
# O comando .apply(criar_frase_marketing) faz o Python rodar nossa funﾃｧﾃ｣o para cada linha da coluna 'nome'
df['News'] = df['Nome'].apply(criar_frase_marketing)

# Mostramos apenas as colunas de Nome e a Mensagem gerada
print("\n--- DADOS TRANSFORMADOS ---")
print(df[['Nome', 'News']])
````

````python
# O comando to_csv exporta o nosso DataFrame para um arquivo real na pastinha do Colab
# index=False serve para nﾃ｣o criar uma coluna de nﾃｺmeros extras (0, 1, 2) no arquivo final
df.to_csv('SDW2023_finalizado.csv', index=False)

print("\n--- CARREGAMENTO CONCLUﾃ好O ---")
print("O arquivo 'SDW2023_finalizado.csv' foi gerado com sucesso!")
````

# Santander Dev Week 2023 (ETL com Python)

**Contexto:** Vocﾃｪ ﾃｩ um cientista de dados no Santander e recebeu a tarefa de envolver seus clientes de maneira mais personalizada. Seu objetivo ﾃｩ usar o poder da IA Generativa para criar mensagens de marketing personalizadas que serﾃ｣o entregues a cada cliente.

**Condiﾃｧﾃｵes do Problema:**

1. Vocﾃｪ recebeu uma planilha simples, em formato CSV ('SDW2023.csv'), com uma lista de IDs de usuﾃ｡rio do banco:
  ```
  UserID
  1
  2
  3
  4
  5
  ```
2. Seu trabalho ﾃｩ consumir o endpoint `GET https://sdw-2023-prd.up.railway.app/users/{id}` (API da Santander Dev Week 2023) para obter os dados de cada cliente.
3. Depois de obter os dados dos clientes, vocﾃｪ vai usar a API do ChatGPT (OpenAI) para gerar uma mensagem de marketing personalizada para cada cliente. Essa mensagem deve enfatizar a importﾃ｢ncia dos investimentos.
4. Uma vez que a mensagem para cada cliente esteja pronta, vocﾃｪ vai enviar essas informaﾃｧﾃｵes de volta para a API, atualizando a lista de "news" de cada usuﾃ｡rio usando o endpoint `PUT https://sdw-2023-prd.up.railway.app/users/{id}`.



````python
# Utilize sua prﾃｳpria URL se quiser ;)
# Repositﾃｳrio da API: https://github.com/digitalinnovationone/santander-dev-week-2023-api
sdw2023_api_url = 'https://sdw-2023-prd.up.railway.app'
````

## **E**xtract

Extraia a lista de IDs de usuﾃ｡rio a partir do arquivo CSV. Para cada ID, faﾃｧa uma requisiﾃｧﾃ｣o GET para obter os dados do usuﾃ｡rio correspondente.

````python
import pandas as pd

df = pd.read_csv('SDW2023.csv')
user_ids = df['UserID'].tolist()
print(user_ids)
````

````python
import requests
import json

def get_user(id):
  response = requests.get(f'{sdw2023_api_url}/users/{id}')
  return response.json() if response.status_code == 200 else None

users = [user for id in user_ids if (user := get_user(id)) is not None]
print(json.dumps(users, indent=2))
````

## **T**ransform

Utilize a API do OpenAI GPT-4 para gerar uma mensagem de marketing personalizada para cada usuﾃ｡rio.

````python
# Documentaﾃｧﾃ｣o Oficial da API OpenAI: https://platform.openai.com/docs/api-reference/introduction
# Informaﾃｧﾃｵes sobre o Perﾃｭodo Gratuito: https://help.openai.com/en/articles/4936830

# Para gerar uma API Key:
# 1. Crie uma conta na OpenAI
# 2. Acesse a seﾃｧﾃ｣o "API Keys"
# 3. Clique em "Create API Key"
# Link direto: https://platform.openai.com/account/api-keys

# Substitua o texto TODO por sua API Key da OpenAI, ela serﾃ｡ salva como uma variﾃ｡vel de ambiente.
openai_api_key = 'TODO'
````

````python
import openai

openai.api_key = openai_api_key

def generate_ai_news(user):
  completion = openai.ChatCompletion.create(
    model="gpt-4",
    messages=[
      {
          "role": "system",
          "content": "Vocﾃｪ ﾃｩ um especialista em markting bancﾃ｡rio."
      },
      {
          "role": "user",
          "content": f"Crie uma mensagem para {user['name']} sobre a importﾃ｢ncia dos investimentos (mﾃ｡ximo de 100 caracteres)"
      }
    ]
  )
  return completion.choices[0].message.content.strip('\"')

for user in users:
  news = generate_ai_news(user)
  print(news)
  user['news'].append({
      "icon": "https://digitalinnovationone.github.io/santander-dev-week-2023-api/icons/credit.svg",
      "description": news
  })
````

## **L**oad

Atualize a lista de "news" de cada usuﾃ｡rio na API com a nova mensagem gerada.

````python
def update_user(user):
  response = requests.put(f"{sdw2023_api_url}/users/{user['id']}", json=user)
  return True if response.status_code == 200 else False

for user in users:
  success = update_user(user)
  print(f"User {user['name']} updated? {success}!")
````
